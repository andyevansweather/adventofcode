```python
import json
```

```python
with open("test.json", "r") as f:
    data = json.load(f)
# data  = [
#   {
#     "forward": 5
#   },
#   {
#     "down": 5
#   },
#   {
#     "forward": 8
#   },
#   {
#     "up": 3
#   },
#   {
#     "down": 8
#   },
#   {
#     "forward": 2
#   }
# ]
```

```python
vert = 0
horiz = 0
depth = 0
for datum in data:
    # print(data[i-1], data[i])
    if "forward" in datum:
        print(datum.get("forward"), vert)
        horiz += datum.get("forward")
        depth += datum.get("forward") * vert
        print("*************")
        print(vert)
        print(horiz)
        print(depth)
        print("*************")
        
    if "down" in datum:
        vert += datum.get("down")
        print("*************")
        print(vert)
        print(horiz)
        print("*************")
        
    if "up" in datum:
        vert -= datum.get("up")
        print("*************")
        print(vert)
        print(horiz)
        print("*************")
        
depth*horiz
```
